```python
import cv2
import numpy as np

# =============== [ P A R A M E T E R ]
#=========================================================
pathResultImages = r"Result_Images\\"
heightBEV = 785
widthBEV = 657
# =============== Camera Filename Images
Camera_A_Name = "6.1.BEV-FINAL-SameSize-Camera_A_1.jpg"
Camera_B_Name = "6.1.BEV-FINAL-SameSize-Camera_B_1.jpg"
Camera_C_Name = "6.1.BEV-FINAL-SameSize-Camera_C_1.jpg"
Camera_D_Name = "6.1.BEV-FINAL-SameSize-Camera_D_1.jpg"
print("=> Filename: ", Camera_A_Name,Camera_B_Name,Camera_C_Name,Camera_D_Name)
print()
# =============== Coordinate Small Square [BEST] Inside
pointXMin = 225
pointXMax = 411
pointYMin = 290
pointYMax = 519
print("=> pointXMin = ", pointXMin)
print("=> pointXMax = ", pointXMax)
print("=> pointYMin = ", pointYMin)
print("=> pointYMax = ", pointYMax)
print("="*50)
print()
# =============== Masking Parameter 
Parameter_Masking_NameX = r"_PARAMS_MASKING_HORIZONTAL_.jpg"
Parameter_Masking_NameY = r"_PARAMS_MASKING_VERTICAL_.jpg"
#=========================================================
# =============== CAR Image (Dimension must [205x189])
Car_Name = "_CAR_.jpg"
#=========================================================
# =============== [ P A R A M E T E R ]

# =============== [ F U N C T I O N ]
#=========================================================
def readImageReal(cameraName, Camera_Filename):
    if cameraName == "Mask":
        img = cv2.imread(Camera_Filename, 0)
    else:
        img = cv2.imread(Camera_Filename)
    '''
    cv2.imshow(Camera_Filename, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    '''
    return img
def stitchCamera(imgA,imgB,imgC,imgD, imgMaskX, imgMaskY):
    cv2.imshow('imgMaskX', imgMaskX)
    cv2.imshow('imgMaskY', imgMaskY)

    # ==================== ADD
    img_add_Ver = cv2.add(imgA,imgD)
    img_add_Hor = cv2.add(imgB,imgC)

    # ==================== BITWISE AND & MASKING [VERTICAL]
    img_add_Ver = cv2.bitwise_and(img_add_Ver,img_add_Ver,mask = imgMaskX)
    # ERASE THE ERROR WITH LINE (BLACK DOTS)
    color = (0,0,0)
    thickness = 10
    img_add_Ver = cv2.line(img_add_Ver,(407,290),(592,0),color,thickness)
    img_add_Ver = cv2.line(img_add_Ver,(227,290),(62,0),color,thickness)
    img_add_Ver = cv2.line(img_add_Ver,(235,519),(84,785),color,thickness)
    img_add_Ver = cv2.line(img_add_Ver,(389,519),(508,785),color,thickness)
    # ==================== BITWISE AND & MASKING [HORIZONTAL]
    #img_add_Hor = cv2.bitwise_and(img_add_Hor,img_add_Hor,mask = imgMaskY)
    cv2.imshow('img_add_Hor', img_add_Hor)
    cv2.imshow('img_add_Ver', img_add_Ver)

    # ==================== ADD - FINAL
    imgStitching = cv2.add(img_add_Ver,img_add_Hor)
    '''
    cv2.imshow('Stitching - Final ', imgStitching)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    '''
    return imgStitching
def addCar(img, imgCar, xMin,xMax,yMin,yMax):
    cv2.imshow('img', img)
    cv2.imshow('imgCar', imgCar)
    img[yMin:yMax, xMin:xMax] = imgCar
    '''
    cv2.imshow('Stitching Car - Final ', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    '''
    return img
def addLine(img):
    x = pointXMax - pointXMin
    x1 = pointXMin + int(x/8)
    x2 = pointXMax - int(x/8)
    xWidth = int(x/7)
    y12 = pointYMax+5
    vertical = 45
    thickness = 2
    # Red Line
    yLast = y12+40
    colorRed = (0,0,255)
    img = cv2.line(img, (x1, y12), (x1, yLast), colorRed, thickness)
    img = cv2.line(img, (x2, y12), (x2, yLast), colorRed, thickness)
    img = cv2.line(img, (x1, yLast), (x1+xWidth, yLast), colorRed, thickness)
    img = cv2.line(img, (x2, yLast), (x2-xWidth, yLast), colorRed, thickness)
    # Green Line
    yLast2 = yLast+45
    colorGreen = (0,255,0)
    img = cv2.line(img, (x1, yLast+5), (x1, yLast2), colorGreen, thickness)
    img = cv2.line(img, (x2, yLast+5), (x2, yLast2), colorGreen, thickness)
    img = cv2.line(img, (x1, yLast2), (x1+xWidth, yLast2), colorGreen, thickness)
    img = cv2.line(img, (x2, yLast2), (x2-xWidth, yLast2), colorGreen, thickness)
    # Green Line 1
    yLast3 = yLast2+45
    colorGreen = (0,255,0)
    img = cv2.line(img, (x1, yLast2+5), (x1, yLast3), colorGreen, thickness)
    img = cv2.line(img, (x2, yLast2+5), (x2, yLast3), colorGreen, thickness)
    img = cv2.line(img, (x1, yLast3), (x1+xWidth, yLast3), colorGreen, thickness)
    img = cv2.line(img, (x2, yLast3), (x2-xWidth, yLast3), colorGreen, thickness)
    '''
    cv2.imshow('addLine', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    '''
    return img
def cropParts(img):
    #Crop
    xDistance = 125
    yDistance = 145
    img = img[pointYMin-yDistance : pointYMax+yDistance , pointXMin-xDistance : pointXMax+xDistance]
    # Rescale
    h,w,c = img.shape
    scale = 1.6
    dim = (int(w*scale), int(h*scale))
    img = cv2.resize(img, dsize=dim, interpolation=cv2.INTER_CUBIC)

    cv2.imshow('cropRescaleParts', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return img

#=========================================================
def writeNewImage(img, pathWriteImg):
    cv2.imwrite(pathWriteImg, img)
#=========================================================
# =============== [ F U N C T I O N ]

# =============== [ M A I N - F U N C T I O N ]
#=========================================================
def main():
    print("==> 1. Size BEV")
    print("Height: ", heightBEV)
    print("Width: ", widthBEV)
    print()

    print("==> 2. Read Camera")
    print()
    # ==================== Read Camera
    imgCameraA = readImageReal("A", Camera_A_Name)
    imgCameraB = readImageReal("B", Camera_B_Name)
    imgCameraC = readImageReal("C", Camera_C_Name)
    imgCameraD = readImageReal("D", Camera_D_Name)
    # ==================== Read Masking 
    imgMaskingX = readImageReal("Mask", Parameter_Masking_NameX)
    imgMaskingY = readImageReal("Mask", Parameter_Masking_NameY)
    # ==================== Read Car
    imgCar = readImageReal("Car", Car_Name)

    print("==> 3. Stitching Camera A-B-C-D")
    print()
    # ==================== Stitching Camera A-B-C-D
    imgStitch = stitchCamera(imgCameraA,imgCameraB,imgCameraC,imgCameraD, imgMaskingX,imgMaskingY)

    print("==> 4. Add Car Image")
    print()
    # ==================== Add Car Image
    imgStitch = addCar(imgStitch, imgCar, pointXMin,pointXMax,pointYMin,pointYMax)

    print("==> 5. Line Behind Car")
    print()
    imgStitch = addLine(imgStitch)

    print("==> 6. Crop Unimportant Parts and Rescale")
    print()
    imgStitch = cropParts(imgStitch)

    print("==> 7. Write new Images")
    print()
    # ==================== Write new Images
    filenameImage = "1.STITCHING-BEV-CAR-FINAL-CROP-"
    filenameImage = pathResultImages + filenameImage
    writeNewImage(imgStitch, filenameImage+Camera_A_Name)
#=========================================================
# =============== [ M A I N - F U N C T I O N ]

if __name__ == '__main__':       
    # Calling main() function 
    main()
```
